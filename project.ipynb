# Random Forest

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, precision_recall_curve, auc

# Prepare data
X = df.drop(columns=["Class"])
y = df["Class"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Train model
clf = RandomForestClassifier(n_estimators=100, class_weight="balanced", random_state=42)
clf.fit(X_train, y_train)

# Evaluate
y_pred = clf.predict(X_test)
y_scores = clf.predict_proba(X_test)[:, 1]  # Get probability scores for the positive class

print(classification_report(y_test, y_pred))

# Compute precision-recall curve
precision, recall, _ = precision_recall_curve(y_test, y_scores)
pr_auc = auc(recall, precision)  # AUC-PR score

print(f"Precision-Recall AUC: {pr_auc:.4f}")

# Plot Precision-Recall Curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.', label=f'AUC = {pr_auc:.4f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.grid()
plt.show()
```

# SMOTE + Random Forest

```python
from imblearn.over_sampling import SMOTE

# Separate features and target
X = df.drop(columns=["Class"])
y = df["Class"]

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Apply SMOTE
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Train Random Forest with SMOTE data
clf_smote = RandomForestClassifier(n_estimators=100, class_weight="balanced", random_state=42)
clf_smote.fit(X_train_smote, y_train_smote)

# Predict and evaluate
y_pred_smote = clf_smote.predict(X_test)
y_scores_smote = clf_smote.predict_proba(X_test)[:, 1]  # Get probability scores
print(classification_report(y_test, y_pred_smote))

# Compute PR Curve & AUC
precision_smote, recall_smote, _ = precision_recall_curve(y_test, y_scores_smote)
pr_auc_smote = auc(recall_smote, precision_smote)

print(f"SMOTE + RF Precision-Recall AUC: {pr_auc_smote:.4f}")

# Plot Precision-Recall Curve
plt.figure(figsize=(8, 6))
plt.plot(recall_smote, precision_smote, marker='.', label=f'AUC = {pr_auc_smote:.4f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve (SMOTE + RF)')
plt.legend()
plt.grid()
plt.show()
```

# XGBoost

```python
from xgboost import XGBClassifier

# Train XGBoost model
xgb_clf = XGBClassifier(scale_pos_weight=len(y_train[y_train == 0]) / len(y_train[y_train == 1]), random_state=42)
xgb_clf.fit(X_train, y_train)

# Predict and evaluate
y_pred_xgb = xgb_clf.predict(X_test)
y_scores_xgb = xgb_clf.predict_proba(X_test)[:, 1]  # Get probability scores
print(classification_report(y_test, y_pred_xgb))

# Compute PR Curve & AUC
precision_xgb, recall_xgb, _ = precision_recall_curve(y_test, y_scores_xgb)
pr_auc_xgb = auc(recall_xgb, precision_xgb)

print(f"XGBoost Precision-Recall AUC: {pr_auc_xgb:.4f}")

# Plot Precision-Recall Curve
plt.figure(figsize=(8, 6))
plt.plot(recall_xgb, precision_xgb, marker='.', label=f'AUC = {pr_auc_xgb:.4f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve (XGBoost)')
plt.legend()
plt.grid()
plt.show()
```

# Isolation Forest

```python
from sklearn.ensemble import IsolationForest

# Train Isolation Forest
iso_forest = IsolationForest(contamination=0.001, random_state=42)  # Adjust contamination rate
iso_forest.fit(X_train)

# Get anomaly scores (higher scores indicate more likelihood of fraud)
y_scores_iso = -iso_forest.decision_function(X_test)  # Invert scores so higher values mean more fraudulent

# Compute precision-recall curve
precision, recall, _ = precision_recall_curve(y_test, y_scores_iso)
pr_auc_iso = auc(recall, precision)

# Predict anomalies (-1 = fraud, 1 = normal) and convert to binary format
y_pred_iso = iso_forest.predict(X_test)
y_pred_iso = [1 if x == -1 else 0 for x in y_pred_iso]  # Convert: -1 (fraud) → 1, 1 (normal) → 0

# Print classification report
print(classification_report(y_test, y_pred_iso))
print(f"Precision-Recall AUC (Isolation Forest): {pr_auc_iso:.4f}")

# Plot Precision-Recall Curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.', label=f'AUC = {pr_auc_iso:.4f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve (Isolation Forest)')
plt.legend()
plt.grid()
plt.show()
```
